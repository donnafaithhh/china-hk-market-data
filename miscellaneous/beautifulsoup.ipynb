```python
# standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# beaultifulsoup
import requests
from bs4 import BeautifulSoup
```

## Learning BeautifulSoup

### Scraping S&P 500 Companies

```python
# setting up 
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
headers = {
    "User-Agent": "AppleProject/1.0 (github link and email)"
}

# access website
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, 'html.parser')

# get table
table = soup.find("table", id="constituents")
rows = table.find('tbody').find_all('tr')

# get companies
companies = []
for row in rows:
    cols = row.find_all('td')
    # print(row)
    # print()
    if len(cols) == 8:
        company = {
            'Symbol': cols[0].get_text(strip=True),
            'Security': cols[1].get_text(strip=True),
            'GICS Sector': cols[2].get_text(strip=True),
            'GICS Sub-Industry': cols[3].get_text(strip=True),
            'Headquarters Location': cols[4].get_text(strip=True),
            'Date added': cols[5].get_text(strip=True),
            'CIK': cols[6].get_text(strip=True),
            'Founded': cols[7].get_text(strip=True)
        }
        companies.append(company)
    
companies_df = pd.DataFrame(companies)
companies_df.head(3)
```
